## Chapter 2, The Vector

This chapter introduced the vector. There are various ways to think about a vector:
- A list of numbers
- A function. For a finite set D and a field $\mathbb F$, a D-vector over $\mathbb F$ is a function from D to $\mathbb F$. This is the representation that the book generally emphasizes. 
- An arrow. A n-vector over $\R$ can be visualized as arrows in $\R^n$. This representation is useful for gaining an intuition of vector addition and scalar-vector multiplication. 
- A translation. "Go east two miles and north three miles" 

Vectors can represent many things: the state of a system, binary string, attributes, probability distribution, an image, just to name a few.

We can add vectors together. Vector addition is associative and commutative. Vector subtraction is the inverse of vector addition. 

We can multiply vectors by scalars, which intuitively scales them. Scalar vector multiplication is associative. 

We can represent lines as a set of scaled vectors. Expressions in the form $\alpha u + \beta v$ where $\alpha,\beta \geq 0$ and $\alpha + \beta = 1$ is called a convex combination. We were also introduced to affine combinations.

The dot product is the sum of the product of corresponding entries. Examples of some of their applications. 
- Dot product between a vector of rates ($/amount) and a vector of amounts gives the total cost
- Dot product between a n-vector and a n-vector where every component is 1/n is the mean. I quite liked this one. 
- Dot product between the voting records of senators where their voting record can be represented as a vector where yes = 1, no = -1, and abstain = 0. The dot product between their voting records can be a measure of how "similar" two vectors are.
- Dot product between audio vectors gives the similarity between the two audios, to a limited extent. 

We were introduced to linear systems. We learned how to find a solution to a very specific linear system, an upper-triangular system of linear equations, using backward substitution. Backward substitution is quite intuitive, but formalizing it into an algorithm required some thought for me. 

## 2.13 Review Questions
- What is vector addition?

Let $u, v$ be n-vectors over $\mathbb F$. Addition of them is defined in terms of the corresponding  entries $[u_1, \ldots u_n] + [v_1, \ldots v_n] = [v_1+u_1, \ldots v_n+u_n]$. The vectors must be of the same length. 

- What is the geometric interpretation of vector addition? 

Let's say $u$ and $v$ are being added. Vector addition can be visualized as putting the tail of $v$ at the head of $u$. The sum is a new arrow drawn from the tail of $u$ to the head of $v$. 

- What is scalar vector multiplication?

Scalar vector multiplication is the multiplication of a scalar by a vector. Every component of the vector gets multiplied by the scalar. $\alpha v = [\alpha v_1, \alpha v_2, \ldots \alpha v_n]$

- What is the distributive property that involves scalar multiplication and but not vector addition?

$(\alpha + \beta) v = \alpha v + \beta v$

- What is the distributive property that involves both scalar-vector multiplication and vector addition? 

$\alpha (u + v) = \alpha u + \alpha v $

- How is scalar-vector multiplication used to represent the line through the origin and a given point?

A line between the origin an a given point can be represented as $[\alpha v_1, \alpha v_2], \alpha \in [0,1]]$. 

Another representation is the convex combination. For a line that goes through $[v_1, v_2]$ and $[u_1, u_2]$, the line between them can be written as $\{\alpha [v_1, v_2] + \beta [u_1, u_2] : \alpha, \beta \in \R, \alpha, \beta \geq 0, \alpha + \beta = 1\}$


- What is dot-product?

The dot product between n-vectors $u$ and $v$ can be defined as $u \cdot v = \sum _0^{n-1} u_i v_i$

- What is the homogeneity property that relates dot-product to scalar-vector multiplication? 

$(\alpha v) \cdot u = \alpha (v \cdot u) $

- What is the distributive property property that relates dot-product to vector addition? 

$(\alpha + \beta)\cdot v = \alpha v \cdot \beta v$


- What is a linear equation? 

A linear equation takes the from $a \cdot c = \beta$

- What is a linear system? 

A linear system is a collection of equations involving one or more of the same variable. The book doesn't provide the most explicit definition. 

- What is an upper-triangular linear system? 

An upper-triangular linear system is a linear system with $n$ equations where equation 1 has no zeros, equation 2 has one zero in the first position, and so on until the $n$th equation which has all zeros except the last position. 

- How can one solve an upper-triangular linear system? 

It can be solved using backward substitution.
